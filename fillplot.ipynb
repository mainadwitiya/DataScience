```
!pip install items
import items
df=pd.read_csv(io.BytesIO(uploaded['salaries.csv']))

```

```
ages=df['Age']
all_dev=df['All_Devs']
python_dev=df['Python']
javascript=df['JavaScript']

plt.plot(ages,all_dev,color='#444444',linestyle='--',label='All devs')
plt.plot(ages,python_dev,label='Python')
plt.style.use('seaborn')
overall_median=57287
#overall median is average
plt.fill_between(ages,python_dev,overall_median,where=(python_dev>overall_median),interpolate=True,alpha=.25)
plt.fill_between(ages,python_dev,overall_median,where=(python_dev<=overall_median),interpolate=True,color='Red',alpha=.25)
 
plt.title('Median Salary in USA by Age')
plt.xlabel('Ages')
plt.ylabel('Salary in USA')
plt.legend()
plt.show()
```
